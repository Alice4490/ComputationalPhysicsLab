```julia
using LinearAlgebra
```

```julia

function QR_decomposition(A::AbstractMatrix)
    m, n = size(A)
    Q = zeros(m,n)
    R = zeros(n,n)
    
    for j in 1:n
        v = A[:, j]

        for i in 1:j-1
            R[i, j] = (Q[:, i]' * v)
            v = v - Q[:, i] * (Q[:, i]' * v)
            
        end
        Q[:, j] = v / norm(v)
        R[j, j] = (Q[:, j]' * v)
    end

    return Q, R
end


function trasporta(A)
    m, n = size(A)
    AT = zeros(n, m)
    for i in 1:m
        for j in 1:n
            AT[j, i] = A[i, j]
        end
    end
    return AT
end
```

```julia
A = [
    2   -1
    0    1
    -2   2
    ]

Q, R = QR_decomposition(A)
println("Matrice Q:")
println(Q)
println("Matrice R:")
println(R)
println("Verifica QR = A:")
println(Q * R)
println(trasporta(Q) * Q)  # Verifica che Q sia ortogonale
```

```julia
B = [
    2   3   4
    4   5   10
    4   8   2
    ]

Q, R = QR_decomposition(B)
println("Matrice Q:")
println(Q)
println("Matrice R:")
println(R)
println("Verifica QR = B:")
println(Q * R)
println(trasporta(Q) * Q)
```

```julia
C = [
    2   -1   4
    0   3    5
    7   2   -6
    1  -4    0
    ]

Q, R = QR_decomposition(C)
println("Matrice Q:")
println(Q)
println("Matrice R:")
println(R)
println("Verifica QR = C:")
println(Q * R)
println(trasporta(Q) * Q)
```
