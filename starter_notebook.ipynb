## 1. Hello World

Let's start with a simple "Hello World" example:

```c++
#include <iostream>

std::cout << "Hello, Computational Laboratory!" << std::endl;
```

## 2. Variables and Basic Operations

Working with variables in C++:

```c++
// Declare and initialize variables
int a = 10;
int b = 20;
int sum = a + b;

std::cout << "a = " << a << std::endl;
std::cout << "b = " << b << std::endl;
std::cout << "sum = " << sum << std::endl;
```

## 3. Functions

Define and use functions:

```c++
// Define a function to calculate factorial
int factorial(int n) {
    if (n <= 1) return 1;
    return n * factorial(n - 1);
}

// Test the function
std::cout << "Factorial of 5 = " << factorial(5) << std::endl;
std::cout << "Factorial of 7 = " << factorial(7) << std::endl;
```

## 4. Loops and Iterations

Working with loops:

```c++
// Print first 10 numbers
std::cout << "First 10 numbers:" << std::endl;
for (int i = 1; i <= 10; i++) {
    std::cout << i << " ";
}
std::cout << std::endl;
```

## 5. Arrays and Vectors

Working with arrays and vectors:

```c++
#include <vector>

// Create a vector
std::vector<int> numbers = {1, 2, 3, 4, 5};

// Print vector elements
std::cout << "Vector elements: ";
for (int num : numbers) {
    std::cout << num << " ";
}
std::cout << std::endl;

// Add an element
numbers.push_back(6);
std::cout << "After adding 6: size = " << numbers.size() << std::endl;
```

## Tips for Using Jupyter with C++

1. **Include headers once**: You only need to include headers once in the notebook (they persist across cells)
2. **Functions and variables persist**: Once defined, functions and variables are available in subsequent cells
3. **Error messages**: If you get an error, read the error message carefully - it will help you debug
4. **Restart kernel**: If things get messy, you can restart the kernel from the "Kernel" menu
5. **Save often**: Remember to save your work frequently (Ctrl+S or Cmd+S)
