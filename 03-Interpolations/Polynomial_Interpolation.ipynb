```julia
using LinearAlgebra
include("LinearSystems.jl")
import .LinearSystems as ls
```

```julia
for j in 4:4
    n = 4*j
    xᵢ = zeros(n+1)
    y = zeros(n+1)

    h = 2/n
    for i in 1:n+1
        xᵢ[i] = -1+(i-1)*h
    end

    V = hcat([xᵢ.^k for k in 0:n]...)
    println("matrice",V)
    k(V) = ls.norma_1(V)*ls.norma_1(inv(V))
    log_k = log(k(V))
    println("n = $n, condizione della matrice di Vandermonde (log): $log_k")
    for l in 1:10
        k = l*10
        function f(x)
            return cos(k*x)
        end
        for i in 1:n+1
            y[i] = f(xᵢ[i])
        end
        println("vettore",y)
        coefficienti = ls.row_pivoting_solution(V, y)
        display(coefficienti)
        termine_noto = ls.row_pivoting_solution(V, y)[1]
        println("n = $n, k = $k, coefficiente di grado 0: $termine_noto")
    end
end
```

```julia
# Funzione per calcolare i polinomi di Legendre fino a grado n in x
function legendre_polynomials(n, x)
    P = zeros(n+1)
    P[1] = 1.0
    if n >= 1
        P[2] = x
        for k in 2:n
            P[k+1] = ((2k - 1)*x*P[k] - (k-1)*P[k-1]) / (k)
        end
    end
    return P
end

# Costruzione della matrice P_n per dato n
function matrix_Plegendre(n, x)

    Pmat = zeros(n + 1, n + 1)
    for i in 1:(n + 1)
        Pmat[i, :] = legendre_polynomials(n, x[i])
    end
    return Pmat
end
x = 1/3
n = 2
legendre_polynomials(n, x)
```

```julia
for n in 4:20
    h = 2/n
    x = zeros(n+1)
    for i in 1:n+1
        x[i] = -1+(i-1)*h
    end
    P = matrix_Plegendre(n, x)
    #println(V)
    kP = ls.norma_1(P)*ls.norma_1(inv(P))
    #log_k = log(k(P))
    println("n = $n, condizione della matrice di Vandermonde: $kP")
end
```
