## Example 2: Basic Variables and Operations

````c
#include <stdio.h>

int main() {
    int a = 10;
    int b = 20;
    int sum = a + b;
    
    printf("Sum of %d and %d is %d\n", a, b, sum);
    return 0;
}
````

## Example 3: Loops

````c
#include <stdio.h>

int main() {
    for (int i = 1; i <= 5; i++) {
        printf("Iteration %d\n", i);
    }
    return 0;
}
````
